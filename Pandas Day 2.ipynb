```python
data={
    "department":["IT","Finance","HR","IT","Finance","HR","IT"],
    "employee":["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "George"],
    "salary":[50000,6500,55000,64000,76000,90000,45000],
    "bonus":[5000, 6000, 7000, 6500, 5500, 6000, 7200]
}
df=pd.DataFrame(data)
print (df)
```

```python
#average salary per department
df.groupby("department")["salary"].sum()
```

```python
# Total salary per department and bonus level
df.groupby(["department","bonus"])["salary"].sum()
```

```python
df.groupby(["bonus","department"])["salary"].sum()
```

```python
# Multiple statistics at once
df.groupby("department")["salary"].agg(["max","min","mean","count"])

#agg ka full form hai aggregate.
```

```python
# Keep only departments where average salary > 55k
filtered=df.groupby("department").filter(lambda x:x["salary"].mean()>55000)
filtered
```

```python
# Add a column with average salary per department
df["Dept_Avg_Salary"] = df.groupby("department")["salary"].transform("mean") #.transform("mean") Har group ka average salary nikalta hai.
df
```

```python
df.groupby("department")["salary"].mean()
```

```python
df['total salary'] = df['salary']+df['bonus']
df
```
