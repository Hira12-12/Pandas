```python
                                                      #12. Combination & Wrangling
#1.	Create a summary DataFrame with columns:
#Category, Total_Products, Avg_Price, Total_Stock, Min_Price, Max_Price
summary = data.groupby("Category").agg(
    Total_Products=("Name", "count"),
    Avg_Price=("Price", "mean"),
    Total_Stock=("Stock", "sum"),
    Min_Price=("Price", "min"),
    Max_Price=("Price", "max")
).reset_index()
summary
```

```python
#2.	Find the ratio of in_stock to pre_order products per category.
availability_ratio = (
    data.groupby(["Category", "Availability"])["Name"]
    .count()
    .unstack(fill_value=0)
)
availability_ratio["ratio_instock_preorder"] = (
    availability_ratio.get("in_stock", 0) / availability_ratio.get("pre_order", 1)
)
availability_ratio 
```

```python
#3. For each brand: number of products, total stock, average price
brand_summary = data.groupby("Brand").agg(
    num_products=("Name", "count"),
    total_stock=("Stock", "sum"),
    avg_price=("Price", "mean")
).reset_index()
brand_summary
```
